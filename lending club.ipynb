## Read Data

```python
data = pd.read_csv('loan.csv')
```

```python
data
```

## Check the percentage of null values in the data

```python
data.isnull().sum()* 100 / len(data)
```

## As we can see, there are some columns where there are 100% null values, we can remove these columns

```python
data = data.dropna(axis=1, how='all')
```

```python
data.isnull().sum()* 100 / len(data)
```

## There are still some columns like next_pymnt_d, mths_since_last_record, mths_since_last_delinq which mostly have null values, we can remove these

```python
data = data.drop(['next_pymnt_d', 'mths_since_last_record', 'mths_since_last_delinq'], axis = 1) 
```

```python
# Number of columns left
len(data.columns)
```

## We can also find number of unique values in each column

```python
data.nunique()
```

## There are columns with only one unique value, these will not play any part in our analysis. So these can be dropped

```python
col = data.nunique()
data = data.drop(list(col[col == 1].index), axis = 1) 
```
